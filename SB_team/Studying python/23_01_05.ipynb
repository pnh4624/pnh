### 4. NumPy ###    
#### 4.1 NumPy package 사용 ####  
- import numpy as np  
- 리스트에서 ndarray 생성: np.array(list)  
cf) list: 각 원소가 각자의 자료형을 가짐, ndarray: 각 원소가 같은 종류의 자료형을 가짐  
ex) b = np.array(list)  
- 행렬의 차원: b.ndim  
- 행렬의 행 수: b.shape   
- 행렬의 인덱싱: b[0,0]  

#### 4.2 특수한 배열의 사용 ####  
- 0부터 1씩 증가하는 원소 10개짜리 배열: np.arange(시작,끝,간격)  
- N_start부터 N_end 전까지 1씩 증가하는 배열: np.arange(N_start, N_end)  
- 크기만 정해진 행렬: np.zeros((2,2)), np.zeros(shape,dtype,order) 
- 유닛 행렬: np.ones((2,3))  
- 모든 원소가 N인 2 X 3 행렬: np.full((2,3), N)  
- 빈 행렬: arr=np.empty(N)
- 단위 행렬: np.eye(n,m,k)  && np.identity(3) 
- 차원 변환: a.reshape((4,5))  && a.flatten()

#### 4.3 Numpy 슬라이싱/인덱싱 ####  
- 행을 0부터 1까지 && 열을 0부터 2까지: arr[0:2, 0:2]  
- 행을 1부터 끝까지 && 열을 1부터 끝까지: arr[1:, 1:]  

#### 4.4 Numpy 연산 ####
- 배열의 사측연산: + - * /  
cf) tuple = (,,,), list = [,,,], array = [[list1],[list2],,]  
arr =  행렬 = [[arr1],[arr2],[arr3]]  
- 배열의 곱: dot(arr1,arr2)  

- 배열의 원소  
 모든 원소의 합: sum()  
 모든 원소의 곱: prod()  
 행 별/열 별 연산: a.sum(axis=0/1)  
 평균: a.mean()  
 분산: a.std()  
 곱: a.prod()  
 +) min, max ...
 
 대각선 원소의 추출: np.diag(arr)  
 
 - 배열의 변환  
hstack, vstack, dstack, stack, r_, c_, tile

 - array의 차원  
 0차원: scalar 1차원: vector 2차원: matrix 3: 3-tensor  
 


```python
import numpy as np
lst = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
arr = np.array(lst)
print(arr)
print()
a = arr[0:3, 0:2]
print(a)

print()

b = arr[1:, 1:]
print(b)
```

### 4.5 난수의 생성 ###
- 시드 설정: np.random.seed(0)  
- 난수 만들기: np.random.rand(N)  
rand: 0부터 1사이의 균일 분포  
randn: 표준 정규 분포  
randint(low, high, size=None): 균일 분포의 정수 난수  

- 순서 바꾸기: np.random.shuffle(array)  

- 샘플링(표본 추출): numpy.random.choice(array, size=None, replace=True, p=None)  
size: 샘플 숫자  
replace: True이면 한번 선택한 데이터를 다시 선택 가능  
p : 배열의 각 원소가 선택될 수 있는 확률(확률 조작)  

- 카운팅
중복 없이 값 골라줌: np.unique(array)  
0까지 쿠운트 해줌: np.bincount(array, minlength=6)
